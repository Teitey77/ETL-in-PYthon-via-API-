```python
# Pagination (adding multiple pages)


def main_request_p(baseurl, endpoint, x):
    r = requests.get(baseurl + endpoint + f'?page={x}')
    return r.json()
# data = main_request_p(baseurl, endpoint, 1) # when you change the number everything will change as well but make it dynamic

# 2
#def get_pages(response):
#    pages = response['info']['pages']
#    return pages
##OR
def get_pages(response):
    return response['info']['pages']
#print(get_pages(data)) #number of pages is 42


# 3
# NOTE: This returns elements at all index through loop
def parse_all_rows(response): 
    charlist = []
    for item in response['results']:
        # print(item['name'], len(item['episode'])) # save this in a dictionary see below. Then save it in a list
        char = {
            'ID': item['id'],
            'name': item['name'],
            'no_ep': len(item['episode'])
        }
        charlist.append(char)
    return charlist
#print(parse_all_rows(data))


# 4
# gives you all pages
mainlist = []
data = main_request_p(baseurl, endpoint, 1) # 1 =page 1
#print(data)
for i in range(1, get_pages(data)+1):
    #print(i) #gives you the number of pages
    data_with_pages = main_request_p(baseurl, endpoint, i)
    #print(data_with_pages) # returns all information_charcter(info & results) for or each page (page 1, 2, 3...)
    desired_info = parse_all_rows(data_with_pages) #parse the json data to filter for the desired information agin for each page ( page 1, 2, 3...)
    mainlist.extend(desired_info) # after each page iteration, it stores the pages out in the lsit then it extend it to the other page ...reptitive

#print(mainlist)

df = pd.DataFrame(mainlist)
print(df.head(), df.tail())
df


```
